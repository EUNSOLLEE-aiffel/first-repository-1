#사각형 넓이 구하기

```python
class Square:
    def __init__(self): #Square 함수를 만들겠다
        self.square = int(input('넓이를 구하고 싶은 사각형의 숫자를 입력해 주세요, 1.직사각형, 2.평행사변형, 3.사다리꼴'))#사각형의 종류를 선택

        if self.square == 1:
            print('직사각형의 함수는 rect()입니다.')
            print(self.rect())
        elif self.square == 2:
            print('평행사변형의 함수는 par()입니다.')
            print(self.par())
        elif self.square == 3:
            print('사다리꼴의 함수는 trape()입니다.')
            print(self.trape())
        else:
            print('숫자 1,2,3으로 다시 입력하세요')
            return Square()

    #직사각형 구하는 공식
    def rect(self):
        width, vertical = map(float, input('가로, 세로를 띄어쓰기로 구분해서 적어주세요').split())
        area = width*vertical #가로*세로
        result = '직사각형의 넓이는 : ' + str(area) #str인 result랑 합치기 위해서 area결과값을 str로 바꿔줌
        return result

    #평행사변형 구하는 공식 입니다.
    def par(self):
        width, vertical = map(float, input('가로, 세로를 띄어쓰기로 구분해서 적어주세요').split())
        area = width*vertical #가로*세로
        result = '평행사변형 넓이는 : ' + str(area) #str인 result랑 합치기 위해서 area결과값을 str로 바꿔줌
        return result
    #사다리꼴 구하는 공식 입니다.
    def trape(self):
        upwidth, downwidth, vertical =  map(float, input('윗변, 아래변, 세로를 띄어쓰기로 구분해서 순서대로 적어주세요').split())
        area = (abs(upwidth-downwidth)/2)*vertical
        result = '사다리꼴 넓이는 : ' + str(area) #str인 result랑 합치기 위해서 area결과값을 str로 바꿔줌
        return result

calc = Square()
#회고 정말 여러가지 방법을 써서 시도하면서 calc.rect, calc.par, calf.trape를 쓸려고 했는데 결국 실패 했습니다.
```

#키오스크 만들기

```python
#키오스크 만들기
class Kiosk:
    def __init__(self):
        '''class Kiosk를 init하는 함수'''
        self.menu = ['americano', 'latte', 'mocha', 'yuza_tea', 'green_tea', 'choco_latte']
        self.price = [2000, 3000, 3000, 2500, 2500, 3000]

    # 메뉴 출력 메서드
    def menu_print(self):
        for i in range(len(self.menu)):
            print(i+1, self.menu[i], ' : ', self.price[i], '원')

    # 주문 메서드
    def menu_select(self):
        '''주문 리스트를 만들고 저장하는 함수입니다.'''
        self.order_menu = []  # 주문 리스트
        self.order_price = []  # 가격 리스트
        n = 0 #n=0으로 넣어서 while이 처음은 무조건 실행
        while n < 1 or len(self.menu) < n:
            n = int(input("음료를 번호를 입력하세요 : "))  # 음료 번호 입력

            # 메뉴판에 있는 음료 번호일 때
            if 1 <= n & n <= len(self.menu):
                self.order_price.append(self.price[n-1])  # 여기 두개는 잘 작동이 안되서 쓰지 못하고 밑에서 다시 append생성
                self.price_sum = self.price[n-1]  #
            # 메뉴판에 없는 번호일 때
            else:
                print("없는 메뉴입니다. 다시 주문해 주세요.")

        # 음료 온도 물어보기
        t = 0  # t=0으로 둬서 while을 무조건 시작
        while t != 1 and t != 2:  # 1이나 2를 입력할 때까지 물어보기
            t= int(input("HOT 음료는 1을, ICE 음료는 2를 입력하세요 : "))
            # 문제 2-3의 답을 입력하세요.
            if t == 1:
                self.temp = 'HOT'
            elif t == 2:
                self.temp = 'ICE'
            else:
                print("1과 2 중 하나를 입력하세요.\n")

        self.order_menu.append(self.temp + ' ' + self.menu[n-1])  # 주문 리스트에 추가합니다.

        # 문제 2-2의 답을 입력하세요.
        print(self.temp, self.menu[n-1], ' : ', self.price[n-1], '원')  # 온도 속성을 추가한 주문 결과를 출력합니다.

        # 추가 주문 또는 지불
        while n != 0:  # 지불을 선택하기 전까지 반복합니다.
            print()  # 줄 바꾸면서
            n = int(input("추가 주문은 음료 번호를, 지불은 0을 누르세요 : "))  # 추가 주문 또는 지불
            if n > 0 and n < len(self.menu) + 1:
                # 추가 음료 온도
                # 문제 2-4. 추가 음료의 온도를 입력받고, 가격 리스트, 주문 리스트, 합계 금액을 업데이트해보세요.
                t = 0
                while t != 1 and t != 2:
                    t= int(input("HOT 음료는 1을, ICE 음료는 2를 입력하세요 : "))
                    if t == 1:
                        self.temp = 'HOT'
                    elif t == 2:
                        self.temp = 'ICE'
                    else:
                        print("1과 2 중 하나를 입력하세요.\n")
                self.order_menu.append(self.temp + ' ' + self.menu[n-1]) #추가한 메뉴 list에 저장
                self.order_price.append(self.price[n-1]) #추가주문 가격을 list에 저장
                print('추가 주문 음료', self.temp, self.menu[n-1], ' : ', self.price[n-1], '원\n', '합계 : ', self.order_price, '원')
            else :
                if n == 0 :  # 지불을 입력하면
                    print("주문이 완료되었습니다.")
                    print(self.order_menu, self.order_price)  # 확인을 위한 리스트를 출력합니다.
                else :  # 없는 번호를 입력할 때
                    print("없는 메뉴입니다. 다시 주문해 주세요.")
    #현금과 카드 계산
    def pay(self):
        n = 0
        while n != 1 and n != 2:
            n = int(input("현금(cash):1 카드(card):2 "))
            if n == 1:
                print('직원을 호출하겠습니다')
            elif n == 2:
                print('IC칩 방향에 맞게 카드를 꽂아주세요.')
            else:
                print('1 or 2중에 입력하세요')

    #영수증 생성
    def table(self):
        print('⟝' + '-' * 30 + '⟞')
        for i in range(5):
            print('|' + ' ' * 31 + '|')

        # 주문 상품명 : 해당 금액
        for i in range(len(self.order_menu)):
            print(self.order_menu[i], ' : ', self.order_price[i])

        print('합계 금액 :', sum(self.order_price))

        # 외곽
        for i in range(5):
            print('|' + ' ' * 31+ '|')
        print('⟝' + '-' * 30 + '⟞')



a = Kiosk()  # 객체 생성
a.menu_print()  # 메뉴 출력
a.menu_select()  # 주문
a.pay() #카드 or 캐쉬
a.table() #영수증

#회고 : VS code와 노드를 왔다갔다하면서 들여쓰기가 안맞어서 힘들었다. 노드에서 되는데 VS에선 안되고 VS에선 되는데 노드에선
#       안되는 일이 많았다. VS코드 중간에 colab처럼 마크다운을 만들수 있는 방법이 있을까요?
```
