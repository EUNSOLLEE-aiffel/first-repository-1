```python
import random as rd 

class Account:
    #account를 만들기위한 random 함수
    a = str(rd.randint(100,1000))
    b = str(rd.randint(10,100))
    c = str(rd.randint(100000,1000000))
    d = [a,b,c]#랜덤 계좌의 앞 3자리 중간 2자리 뒤 6자리를 만들기 위해서 나눠서 생성
    '''brand와 account를 설정'''
    brand = 'SC bank'
    account_name = '-'.join(d) #계좌번호 생성
    account_num = 0 #처음 계좌갯수
    account_owners = [] #계좌를 만들면 리스트 생성
    account_balance = []
    num = {}#계좌 개설한 오너와 잔고를 딕셔너리
    account_deposit = 0 #이자를 주기위해서 deposit의 횟수를 카운트
    account_history = 0 #입금과 출금 이자지급의 횟수를 동시에 카운트
    history = [] 

    def __init__(self,name,balance):
        self.name = name #이름 생성
        self.balance = balance #잔고 생성
        Account.account_num +=1 #계좌 한개를 만들때마다 갯수증가

        '''계좌를 개설하는 사람들의 list 저장'''
        Account.account_owners.append(self.name) #계좌 개설한 오너들의 리스트
        Account.account_balance.append(self.balance)#계좌 개설한 잔고
        self.num = {x:y for x,y in zip(Account.account_owners,Account.account_balance)} #계좌 개설한 오너와 잔고를 딕셔너리
        Account.num = self.num #초기값을 위에 num에게 주기위해서 

    #account의 갯수를 세어주는 함수    
    def get_account_num(self): 
        print('현재 생성된 계좌의 갯수는',Account.account_num,'개 입니다.')
   
    #입금을 하는 함수
    def deposit(self):
        a = int(input(f"{self.name}님 얼마를 입금하시겠습니까? 최소 1원이상"))
        if a > 1 :
            self.balance += a
            self.account_deposit +=1 # deposit을 할때마다 1씩 +가 되는구조
            self.account_history +=1 #입금 출금 이자지급 횟수를 통합 카운트
            self.history.append(f'{self.account_history}회: 입금\t금액:{a}\t잔액:{self.balance}')
            print("입금이 완료되었습니다. 현재 잔액은",self.balance,'입니다') #print를 쓰면 None을 return하여서 print를 지우고 return을 쓰니 밑에 이자 시스템을 사용못함....
            if self.account_deposit % 5 == 0:
                print("이자가",self.balance*0.1,'만큼 지급되었습니다')
                self.account_history +=1 #입금 출금 이자지급 횟수를 통합 카운트
                self.history.append(f'{self.account_history}회: 이자지급\t금액:{self.balance*0.1}\t잔액:{self.balance+self.balance*0.1}')#이자지급을 합친 잔액이 나와야되므로
                self.balance = self.balance + (self.balance * 0.1)#이자를 10%줍니다. 1%너무 짜네요  
        else :
            print('잘못 입력하셨습니다.')
            return deposit()
        print(self.balance)
        self.num[self.name] = self.balance # banace 잔고를 바꿔주는 함수 이자가 추가 될경우도 update  
        
    
    #출금을 하는 함수
    def withdraw(self):
        a = int(input(f"{self.name}님 얼마를 출금하시겠습니까?"))
        if a < self.balance :
            self.balance -= a
            self.account_history +=1 #입금 출금 이자지급 횟수를 통합 카운트
            self.history.append(f'{self.account_history}회: 출금\t금액:{a}\t잔액:{self.balance}')
            print("현재 잔액은",self.balance,'남았습니다.')
        else : 
            print("오류입니다. 잔액을 확인하세요")
            return self.withdraw()
        self.num[self.name] = self.balance # banace 잔고를 바꿔주는 함수 출금을 할경우 update
    #고객의 정보를 보여주는 함수
    def display_infor(self):
        return(f'은행이름:{Account.brand},예금주:{self.name},계좌번호:{self.account_name},잔액:{self.balance}')
    

    #100만원 이상의 고객 리스트를 검색하는 함수 
    def search_balance(self):
        n = 0
        while n != 1 and n != 2:
            n = int(input("100만원 이상의 고객을 검색하겠습니까? Yes : 1, No : 2 입력하세요:"))
            if n == 1:
                for key,value in Account.num.items():
                    if value >= 1000000:
                        print(key)
            elif n == 2:
                print('검색을 종료하겠습니다.')
            else:
                print("'1' 또는 '2'으로 입력하세요")

    #입금 출금 이자지급 history함수
    def dwr_history(self):
        for i in range(len(self.history)):
            print(self.history[i])
    
 

Park = Account('박종호',10000)
Dong = Account('김동은',10000000)
Yang = Account('양성모',1000000000)
print(Park.deposit())
print(Park.deposit())
print(Park.withdraw())
print(Park.deposit())
print(Park.deposit())
print(Park.deposit())

print(Park.dwr_history())
print(Park.display_infor())
print(Park.search_balance())
```
